# Diagramme mit Matplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3] # x
y = [2, 5, 2] # y(x)

plt.plot(x, y, "r") # "go/go-" -> Punkte/gestrichelte Linie
plt.show()
```

```python
x2 = [-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2, 2.5]
y2 = []
for element in x2:
    y2.append(element * element) # Quadrat berechnen
print(x2)
print(y2)
```

```python
plt.plot(x2, y2, "r")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")
plt.grid() # Raster
plt.show()
```

# Aufgabe!

Zeichne die Funktion x^3 f√ºr x zwischen -3 und 3!

```python
xx = [-3, -2, -1, 0, 1, 2, 3]
yy = []
for element in xx:
    yy.append(element**3)
print(xx)
print(yy)
```

```python
plt.plot(xx, yy)
plt.show()
```

```python
x3 = [-3, -2, -1, 0, 1, 2, 3]
for i in range(-30, 31): # erstelle x Werte von -30 bis 30
    x3.append(i / 10) # teile alle Werte durch 10 (1.1,1.2, etc)   
y3 = []    
for element in x3:
    y3.append(element**3)
print(x3)
print(y3)
```

```python
plt.plot(x3, y3)
plt.show()
```
